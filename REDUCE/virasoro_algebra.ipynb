### 2024-03-30

`l(m)` はVirasoro代数の $L_m$ の交換関係をみたし, 積を取ると自動的にすべての項の積の順序が m が小さい順になるように交換関係が用いられる.

Singular vectors や Kac determinant などの計算が実装されている.

詳しい使い方については以下を見て欲しい.
